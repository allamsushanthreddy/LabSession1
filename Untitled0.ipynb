```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

##General Graph

```python
# from matplotlib.pyplot import title
x = y = z = p = q = r = s = []

x = np.arange(0.1,9,0.1)
y = 2**x
z = (x**3)/2
p = 5*(x**2)
q = x*50

for i in x:
  r = np.append(r, 40*math.sqrt(i))
  s = np.append(s, 400*math.log(i+1, 10))
  
# Create figure
        
fig = plt.figure() 

# Figure size

fig.set_size_inches(12, 8)

plt.title("Graph")
plt.xlabel("n")
plt.ylabel("T(n)")

plt.plot(x, y, label='2^n', linewidth=2)
plt.plot(x, z, label='1/2 * n**3', linewidth=2)
plt.plot(x, p, label='5*n**2', linewidth=2)
plt.plot(x, q, label='50*n', linewidth=2)
plt.plot(x, r, label='40*sqrt(n)', linewidth=2)
plt.plot(x, s, label='400*log(n)', linewidth=2)

plt.margins(x = 0, y = 0)

plt.text(2, 240, '400*log(n)', fontsize = 12)
plt.text(3, 180, '50*n', fontsize = 12)
plt.text(4.5, 140, '5*n**2', fontsize = 12)
plt.text(8, 130, '40*sqrt', fontsize = 12)
plt.text(8.3, 450, '2^n', fontsize = 12)
plt.text(6, 160, '1/2 * n**3', fontsize = 12)

# plt.xlim([0, 12])
# plt.ylim([0, 12])
# plt.legend()
```

##Time Complexity Graph

```python
from scipy.special import factorial

x = y = z = p = q = r = s = []

x = np.arange(0.1,10,0.1)

y = factorial(x)
z = 2**x
p = x**2
q = x*(np.log(x))
r = np.log(x)
s = x

# Create figure
        
fig = plt.figure() 

# Figure size

fig.set_size_inches(12, 8)

plt.title("Time Complexity Graph")
plt.xlabel("n")
plt.ylabel("T(n)")

plt.plot(x, y, label='n!', linewidth=2)
plt.plot(x, z, label='2**n', linewidth=2)
plt.plot(x, p, label='n**2', linewidth=2)
plt.plot(x, q, label='n*log(n)', linewidth=2)
plt.plot(x, r, label='log(n)', linewidth=2)
plt.plot(x, s, label='n', linewidth=2)

plt.margins(x = 0, y = 0)

plt.legend()
plt.ylim([0,50])
plt.text(4,35, 'n!', fontsize = 12)
plt.text(4.5, 35, '2^n', fontsize = 12)
plt.text(5.5, 35, 'n^2', fontsize = 12)
plt.text(6, 15, 'n*log(n)', fontsize = 12)
plt.text(7, 8, 'n', fontsize = 12)
plt.text(7, 3, 'log(n)', fontsize = 12)
```

###f(n) and g(n) comparision

```python
x = y = z = []

x = np.arange(0,10,0.1)

for i in x:
  y = np.append(y, i**2+3*i+2)
  z = np.append(z, i**2)

# Create figure
        
fig = plt.figure() 

# Figure size

fig.set_size_inches(12, 8)

plt.title("Function Comparison Graph")
plt.xlabel("n")
plt.ylabel("T(n)")

plt.plot(x, y, label='n^2+3n+2', linewidth=2)
plt.plot(x, z, label='n^2', linewidth=2)

plt.margins(x = 0, y = 0)

# plt.legend()
plt.ylim([0,50])
plt.text(3,35, 'n^2+3n+2', fontsize = 12)
plt.text(5.5,35, 'n^2', fontsize = 12)
plt.show()
```

###Regular Graph

```python
x = y = z = p = q = r = s = []

x = np.arange(0,10,0.1)

z = x**0.8
p = x
q = x**1.5
r = x**2
s = x**3

for i in x:
  y = np.append(y, np.math.sqrt(i))

print(y.size, x.size)
# Create figure
        
fig = plt.figure() 

# Figure size

fig.set_size_inches(12, 8)

plt.title("Polynomial Graph")
plt.xlabel("n")
plt.ylabel("T(n)")

plt.plot(x, y, label='sqrt(n)', linewidth=2)
plt.plot(x, z, label='n^0.8', linewidth=2)
plt.plot(x, p, label='n', linewidth=2)
plt.plot(x, q, label='n^1.5', linewidth=2)
plt.plot(x, r, label='n^2', linewidth=2)
plt.plot(x, s, label='n^3', linewidth=2)

plt.margins(x = 0, y = 0)

plt.legend()
plt.ylim([0,50])
plt.text(8.5,1.5, 'Sqrt(n)', fontsize = 12)
plt.text(7.5,6, 'n^0.8', fontsize = 12)
plt.text(7, 8, 'n', fontsize = 12)
plt.text(6.5,20, 'n^1.5', fontsize = 12)
plt.text(5.5,35, 'n^2', fontsize = 12)
plt.text(2.7,35, 'n^3', fontsize = 12)
plt.show()
```
